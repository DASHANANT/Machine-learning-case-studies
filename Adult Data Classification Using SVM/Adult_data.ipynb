```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from numpy.linalg import inv,det,multi_dot
from itertools import combinations
from sklearn.model_selection import train_test_split,KFold,cross_val_score
```

```python
d1=pd.read_csv("adult.data",header=None)
d1.columns =['age', 'workclass', 'fnlwgt', 'education', 'educational-num', 'marital',
       'occupation', 'relationship', 'race', 'gender', 'capital gain',
       'capital loss', 'hours per week', 'country', 'income']
```

# Removing Columns where most data is NULL(0):

## Replacing unwanted values with assigning max attribute value

```python
d1["workclass"].value_counts()
```

```python
def change(x):
    if x == ' ?':
        return " Private"
    return x
d1["workclass"] = d1["workclass"].apply(change)
```

```python
d1["workclass"].value_counts()
```

```python
def change(x):
    if x == ' ?':
        return "United-States"
    return x
d1["country"] = d1["country"].apply(change)
```

```python
d1["country"].value_counts()
```

```python
def change(x):
    if x ==" ?":
        return " Prof-specialty"
    return x
d1["occupation"] = d1["occupation"].apply(change)
```

```python
d1["occupation"].value_counts()
```

```python
data = pd.DataFrame(data=d1)
data
```

## Data Encoding 

```python
c1=[]
for i in d1.columns:    
    c1.append(i)
c1
```

```python
c2=['workclass',  'education', 'marital', 'occupation', 'relationship', 'race', 'gender','country', 'income']
for i in c2:    
    column=np.unique(data[i])    
    for k in range(len(column)):
        data=data.replace({i:column[k]}, {i: k}, regex=True)
```
